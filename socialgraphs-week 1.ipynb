## Write your own little notebook to download wikipedia pages based on the video above. Download the source for your 4 favorite wikipedia pages.

```python
import urllib.request

baseurl='https://en.wikipedia.org/w/api.php?'
action="action=query"
title='titles=Rembrandt'
content='prop=revisions&rvprop=content'
dataformat='format=json'

queryurl='{}{}&{}&{}&{}'.format(baseurl,action,content,title,dataformat)
print (queryurl)
response=urllib.request.urlopen(queryurl)
html=response.read()
text=html.decode('utf-8')
# print (text)
# https://en.wikipedia.org/w/api.php?action=query&prop=revisions&rvprop=content&titles=Rembrandt&format=json
# https://en.wikipedia.org/w/api.php?action=query&prop=revisions&rvprop=content&title=Rembrandt&format=json x
```

```python
import json
json.loads(text)
```

```python
FG = nx.Graph()
FG.add_weighted_edges_from([(1,2,0.125),(1,3,0.75),(2,3,1.2),(3,4,0.375)])
for n,nbrs in FG.adj.items():
    for nbr,eattr in nbrs.items():
        wt=eattr['weight']
        if wt < 0.5: print(f"({n},{nbr},{wt:.3})")
```

```python
for (u,v,wt) in FG.edges.data('weight'):
    if wt < 0.5:
        print(f"({u},{v},{wt:.3})")
```

```python

GG=nx.Graph()
for n,nbrs in MG.adjacency():
    for nbr,edict in nbrs.items():
        minvalue = min([d['weight'] for d in edict.values()])
        GG.add_edge(n,nbr,weight=minvalue)
```

```python
G = nx.petersen_graph()
options={'node_color':'black','node_size':100,'width':3}
# 几行几列第几个
subax1=plt.subplot(221)
nx.draw_random(G,**options)
subax2 = plt.subplot(222)
nx.draw_circular(G, **options)
subax3 = plt.subplot(223)
nx.draw_spectral(G, **options)
subax4 = plt.subplot(224)
nx.draw_shell(G, nlist=[range(5,10), range(5)], **options)
```
